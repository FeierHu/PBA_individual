## 2.Rank based CTR model

```python
rctrdf=df.groupby('contentPosition').aggregate({'IsClicked': sum, 'frontPageViewId': len})
```

```python
rctrdf=rctrdf.rename(columns={'IsClicked':'clicks', 'frontPageViewId':'impressions',})
```

```python
rctrdf['P(Cu=1)']=rctrdf.clicks.fillna(0)/rctrdf.impressions
```

```python
rctrdf['P(Cu=0)']=1-(rctrdf.clicks.fillna(0)/rctrdf.impressions)
```

```python
rctrdf
```

## 3.Position-based model

```python
pbmdf=pd.DataFrame((df.groupby('contentPosition').aggregate({'frontPageViewId': len})).frontPageViewId)/(df.frontPageViewId.nunique())
```

```python
pbmdf=pbmdf.rename(columns={'frontPageViewId':'PEr'})
```

```python
pbmdf
```

```python
pbmdf2=df.groupby('contentId').aggregate({'IsClicked':sum , 'frontPageViewId':len})
```

```python
pbmdf2=pbmdf2.rename(columns={'IsClicked':'clicks', 'frontPageViewId':'impressions',})
```

```python
pbmdf2['PAu']=pbmdf2.clicks.fillna(0)/pbmdf2.impressions
```

```python
pbmdf2=pd.DataFrame(pbmdf2.PAu)
```

```python
pbmdf2
```

```python
pbmdf3=df.groupby(['contentId','contentPosition']).aggregate({'IsClicked':sum})
```

```python
pbmdf3
```

```python
pbmdfall=pbmdf3.join(pbmdf, how='outer')
```

```python
pbmdfall=pbmdfall.join(pbmdf2,how='outer')
```

```python
pbmdfall['PCu']=pbmdfall.PEr*pbmdfall.PAu
```

```python
pbmdfall
```

```python
pbmdfall[pbmdfall['PCu']==pbmdfall.PCu.max()]
```

```python
clicksatisfied = clickdf[clickdf["readingTime"] > 10]
clicksatisfied["satisfied"] = 1
clickunsatisfied = clickdf[clickdf["readingTime"] <= 10]
clickunsatisfied["satisfied"] = 0
```
