```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame, Panel
import csv
```

```python
%matplotlib notebook

raw_impressions = pd.read_csv('../data/20170613-20170619_joined_impressions_small.csv.gz',
    sep=',', compression='gzip', parse_dates=['impressionTimestamp', 'clickTimestamp'])
```

```python
raw_impressions
```

```python
df_clean = raw_impressions.dropna(subset=['environmentId', 'frontPageViewId','contentPosition'], how= 'all')
```

```python
%matplotlib notebook
impressions = df_clean\
    .query('contentPosition > 0').assign(contentPosition=lambda _:_.contentPosition.astype(int))\
    .assign(isClicked=lambda _:_.isClicked.fillna(0))
# impressions
```

```python
print("{}/{} clicked/total impressions".format(
    raw_impressions.isClicked.sum(), raw_impressions.shape[0]))
print("{}/{} clicked/total impressions".format(
    impressions.isClicked.sum(), impressions.shape[0]))
```

```python
impressions.to_pickle("../data/processed/joined_impressions.pkl")
```
