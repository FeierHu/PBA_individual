```python
import itertools

flat_metadata=pd.DataFrame(
    [r for r in itertools.chain.from_iterable([(index, tag) for tag in row['categories'].split(',')] 
                                              for index, row in trydf.set_index('impression').iterrows())],
columns=['article_id', 'category']).set_index('article_id')
```

```python
flat_metadata
```

```python
from sklearn.preprocessing import LabelEncoder  

category_encoder = LabelEncoder()

#flat_metadata['encoded_category'] = category_encoder.fit_transform(flat_metadata.category)
flat_metadata = flat_metadata.assign(encoded_category=lambda df: category_encoder.fit_transform(df.category))

flat_metadata
```

```python
list(flat_metadata.index)
```

```python
initial = [1]*len(flat_metadata.index)
```

```python
from scipy import sparse
output = sparse.coo_matrix(
    (initial, 
    (list(flat_metadata.index), list(flat_metadata.encoded_category))))

output
```
